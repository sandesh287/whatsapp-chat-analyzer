```python
# import libraries
import re
import pandas as pd
```

```python
# open and read file
with open("chat_smarak.txt", "r", encoding="utf-8") as f:
    data = f.read()
```

```python
# display data
print(data)
```

<h6>First create a dataframe for the chat, where columns will be: 1. date, time 2. name, message. This project handles 24-hour format time only.</h6>

```python
# regular expressiom pattern from above data
pattern = r'(?:\u200e)?\[(\d{1,2}/\d{1,2}/\d{2,4}),\s(\d{1,2}:\d{2}:\d{2})\]\s(.*?)(?=\n(?:\u200e)?\[|$)'
```

```python
# convert raw chat text into structured data (list of tuples: date, time, message) using regex
matches = re.findall(pattern, data, re.DOTALL)
matches
```

```python
# convert above data into dataframe of 3 columns: date, time and message
df = pd.DataFrame(matches, columns = ["date", "time", "user_message"])
df
```

```python
# merge date and time column into one
df['datetime'] = pd.to_datetime(df['date'] + ' ' + df['time'], dayfirst=True)
```

```python
# removed unnecessary invisible character from message
df['user_message'] = df['user_message'].str.replace('\u200e', '', regex=False)
```

```python
# display merged dataframe
df = df[['datetime', 'user_message']]
df
```

```python
df.head()
```

```python
df.shape
```

```python
df1 = df.copy()
df1
```

```python
# separate users and messages with regex
users = []
messages = []

for message in df1['user_message']:
    entry = re.split(r'([^:]+):\s?', message)
    if len(entry) > 2:  # user name
        users.append(entry[1].strip())
        messages.append(entry[2].strip())
    else:
        users.append('group_notification')
        messages.append(entry[0].strip())

df1['user'] = users
df1['message'] = messages

# remove user_message column
df1.drop(columns=['user_message'], inplace=True)
```

```python
df1
```

```python
df2 = df.copy()
df2.head()
```

```python
# separate without regex
users = []
messages = []

for message in df2['user_message']:

    if ':' in message:
        user, msg = message.split(':', 1)
        users.append(user.strip())
        messages.append(msg.strip())

    else:
        users.append('group_notification')
        messages.append(message.strip())

df2['user'] = users
df2['message'] = messages

df2.drop(columns=['user_message'], inplace=True)
```

```python
# Heatmap of user of active and inactive time interval
```

```python
period = []
for hour in df2[['day_name', 'hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
df2['period'] = period
```
